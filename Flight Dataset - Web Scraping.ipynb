```python
import time  #importing libraries
import selenium
import pandas as pd 
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
s=Service(r"C:\\Users\\Hinal\\Downloads\\chromedriver_win32\\chromedriver.exe")  #opening chrome
driver = webdriver.Chrome(service=s)    
driver.maximize_window()
driver.get("https://www.yatra.com/")  #getting the webpage of olx

name=[] #empty list
source=[] #empty list
dest=[] #empty list
dep=[] #empty list
arr=[] #empty list
duration =[] #empty list
stops=[] #empty list
info=[] #empty list
price=[] #empty list
url=[] #empty list

time.sleep(2)

#the site uses cookies, click on I agree
Iagree = driver.find_element(By.XPATH, "//button[@class='btngdpr']")
Iagree.click()

#click on first location, "New DElhi", in section of popular domestic flight routes 
domdelhi = driver.find_element(By.XPATH, "//div[@class='popular_destination popular-destination-module carousalModuleArea'][1]/ul/li[1]")
domdelhi.click()

#collecting flights from Delhi to popular stops
urldm = driver.find_elements(By.XPATH,"//div[@class='popular_destination popular-destination-module carousalModuleArea'][1]/div/div/div/div[1]/div/a")
for i in urldm[:]:
        url.append(i.get_attribute('href')) 
        
#click on second location, "Mumbai", in section of popular domestic flight routes 
dommum = driver.find_element(By.XPATH, "//div[@class='popular_destination popular-destination-module carousalModuleArea'][1]/ul/li[2]")
dommum.click()
     
#collecting flights from Mumbai to popular stops
urlmum = driver.find_elements(By.XPATH,"//div[@class='popular_destination popular-destination-module carousalModuleArea'][1]/div/div/div/div[1]/div/a")
for i in urlmum[:]:
        url.append(i.get_attribute('href')) 

#click on third location, "BANGALORE", in section of popular domestic flight routes 
dombgl = driver.find_element(By.XPATH, "//div[@class='popular_destination popular-destination-module carousalModuleArea'][1]/ul/li[3]")
dombgl.click()

#collecting flights from BANGALORE to popular stops
urlbnl = driver.find_elements(By.XPATH,"//div[@class='popular_destination popular-destination-module carousalModuleArea'][1]/div/div/div/div[1]/div/a")
for i in urlbgl[:]:
        url.append(i.get_attribute('href')) 
        
#click on fourth location, "HYDERABAD", in section of popular domestic flight routes 
domhyd = driver.find_element(By.XPATH, "//div[@class='popular_destination popular-destination-module carousalModuleArea'][1]/ul/li[4]")
domhyd.click()
     
#collecting flights from HYDERABAD to popular stops
urlhyd = driver.find_elements(By.XPATH,"//div[@class='popular_destination popular-destination-module carousalModuleArea'][1]/div/div/div/div[1]/div/a")
for i in urlhyd[:]:
        url.append(i.get_attribute('href'))
        
#click on fifth location, "Kolkata", in section of popular domestic flight routes 
domkol = driver.find_element(By.XPATH, "//div[@class='popular_destination popular-destination-module carousalModuleArea'][1]/ul/li[5]")
domkol.click()

#collecting flights from Kolkata to popular stops
urlkol = driver.find_elements(By.XPATH,"//div[@class='popular_destination popular-destination-module carousalModuleArea'][1]/div/div/div/div[1]/div/a")
for i in urlkol[:]:
        url.append(i.get_attribute('href')) 
        
#click on sixth location, "Others", in section of popular domestic flight routes 
domot = driver.find_element(By.XPATH, "//div[@class='popular_destination popular-destination-module carousalModuleArea'][1]/ul/li[6]")
domot.click()
     
#collecting flights from Others to popular stops
urlot = driver.find_elements(By.XPATH,"//div[@class='popular_destination popular-destination-module carousalModuleArea'][1]/div/div/div/div[1]/div/a")
for i in urlot[:]:
        url.append(i.get_attribute('href'))        
#we have fetched urls of 60 pages        
        
for j in url:      #fetching data of various flights from each page
    driver.get(i)   
    time.sleep(2)
    
    #click on view more flights (There are 25 flight details on each page after clicking on view more)
    viewmore=driver.find_element(By.XPATH,"//span[@class='flR mr15 view_more']/a")
    viewmore.click()
    
    #fetching name of the flight
    nm=driver.find_elements(By.XPATH,"//div[@class='fs-13 airline-name no-pad col-8']/span")
    for k in nm:
        name.append(k.text)
    
    #fetching the source 
    src=driver.find_elements(By.XPATH,"//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p")
    for k in src:
        source.append(k.text)
    
    #fetching the destination
    dst=driver.find_elements(By.XPATH,"//div[@class='i-b pdd-0 text-left atime col-5']/p[2]")
    for k in dst:
        dest.append(k.text)
    
    #fetching the departure time
    dt=driver.find_elements(By.XPATH,"//div[@class='i-b pr']")
    for k in dt:
        dep.append(k.text)
    
    #fetching the arrival time
    at=driver.find_elements(By.XPATH,"//div[@class='i-b pdd-0 text-left atime col-5']/p[1]")
    for k in at:
        arr.append(k.text)
    
    #fetching the duration
    dur=driver.find_elements(By.XPATH,"//div[@class='stop-cont pl-13']/p")
    for k in dur:
        duration.append(k.text) 
    
    #fetching the number of stops
    stp=driver.find_elements(By.XPATH,"//div[@class='stop-cont pl-13']/div")
    for k in stp:
        stops.append(k.text)
        
    try:    
        #fetching the additional information of the flight
        ad=driver.find_elements(By.XPATH,"//span[@class='text']")
        for k in ad:
            info.append(k.text)
    except:
        info.append('No info')
    
    #fetching the price 
    prc=driver.find_elements(By.XPATH,"//div[@class='i-b tipsy fare-summary-tooltip fs-18']/i")
    for k in prc:
        price.append(k.text)    

print(len(name),len(source),len(dest),len(dep),len(arr),len(duration),len(stops),len(info),len(price))                      
```

```python
Flight = pd.DataFrame({})   #creating a datafrmae
Flight['Airline']=name
Flight['Source']=source
Flight['Destination']=dest
Flight['Dep_Time']=dep
Flight['Arrival_Time']=arr
Flight['Duration']=duration
Flight['Total_Stops']=stops
Flight['Additional_Info']=info
Flight['Price (in ₹)']=price
Flight  
```
